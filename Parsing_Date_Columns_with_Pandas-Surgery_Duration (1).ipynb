```python
def agg_count(df, group_field):
    grouped = df.groupby(group_field, as_index = True).size().sort_values(ascending = False)
    
    grouped = pd.DataFrame(grouped).reset_index()
    grouped.columns = [group_field, 'Count']
    return grouped
```

```python
category_counts = agg_count(df, 'Institution')

print(category_counts)
```

```python
ax = sns.barplot(data=category_counts, x='Count', y='Institution')

ax.set(xlabel='\n Duration')
ax.set_title('Total OR Time \n')
```

```python
def agg_avg(df, group_field, calc_field):
    grouped = df.groupby(group_field, as_index=False) [calc_field].mean().sort_values(calc_field, ascending = False)
    grouped.columns = [group_field, 'Avg ' + str (calc_field)]
    return grouped
```

```python
avg_dur = agg_avg(df, 'Institution', 'Procedure Duration')

print(avg_dur)
```

```python
ax = sns.barplot(data=avg_dur, x='Avg Procedure Duration', y='Institution')
ax.set(xlabel='\n Average Procedure Duration')
ax.set_title('Average Procedure Duration by MTF \n')
```

```python
def pivot_count(df, rows, columns, calc_field):
    df_pivot = df.pivot_table(values = calc_field,
                             index = rows,
                             columns = columns,
                             aggfunc = np.size
                             ).dropna(axis = 0, how = 'all')
    return df_pivot
```

```python
dur_pivot = pivot_count(df, 'Procedure Duration',
                              'Institution', 'Preop Diagnosis')

print(dur_pivot)
```

```python
sns.heatmap(dur_pivot, annot = True, fmt = 'g')
ax.set(xlabel = '\n Insitution')
ax.set_title('Procedure Duration by Institution \n')
```

## Visualizing changes over time

```python
def multi_line(df, x, y):
    ax = df.groupby([x, y]).size().unstack(y).plot(figsize = (15,8), cmap = 'Set2')
```

```python
multi_line(df, 'Start Date', 'Institution')
ax.set(xlabel='\n Date')
ax.set_title('Institutions Over Time \n')
```
